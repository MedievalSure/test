```python
def fib(max):
    n,a,b=0,0,1
    while n<max:
        yield b
        a,b,n=b,a+b,n+1
```

```python
fib(6)
```

```python
def lazy_sum(*args):
    def sum():
        ax=0
        for n in args:
            ax=ax+n
        return ax
    return sum
```
